````python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from requests.exceptions import HTTPError
````

## Verificando o Robots.txt

````python


requerimento = None
URL = 'https://github.com/trending/robots.txt'
try:
    requerimento = requests.get(URL)
    requerimento.raise_for_status()

except HTTPError as exc:
  print(exc)

else:
    requerimento = requerimento.text
print(requerimento)
````

## Trazendo os dados

````python
requerimento = None
URL = 'https://github.com/trending'
try:
    requerimento = requests.get(URL)
    requerimento.raise_for_status()

except HTTPError as exc:
  print(exc)

else:
    requerimento = requerimento.text
print(requerimento)
````

````python
pagina = BeautifulSoup(requerimento, 'html.parser')

print(pagina.prettify())
````

### Filtrando apenas os dados dos projetos

````python
projetos = pagina.find_all('article',class_='Box-row')[:10]
````

````python
top_projetos = []

for projeto in projetos:

  #titulo dos projetos
  titulo = projeto.find('h2', class_='h3 lh-condensed').text.strip()
  titulo = titulo.replace('\n','').replace('\t','').replace('/', ',').replace('     ', '')

  #linguagem de programação
  linguagem = projeto.find('span', class_='d-inline-block ml-0 mr-3').text.strip()

  #estrelas totais
  stars = projeto.find('a', class_='Link Link--muted d-inline-block mr-3').text.strip()

  #estrelas hoje
  stars_today = projeto.find('span', class_='d-inline-block float-sm-right').text.strip()
  stars_today = stars_today.replace('stars today', '')


  #forks
  forks = projeto.find('a', class_='Link Link--muted d-inline-block mr-3').text.strip()

  top_projetos.append([titulo, linguagem, stars, stars_today, forks])
````

## Convertendo em um DataFrame

````python
df_projeto = pd.DataFrame(top_projetos, columns = ['Título', 'Linguagem', 'Estrelas Hoje', 'Estrelas Totais', 'Forks'])
````

````python
df_projeto['Ranking'] = range(1,11)
````
